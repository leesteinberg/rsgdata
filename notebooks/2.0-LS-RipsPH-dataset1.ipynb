# Persistent Homology (Rips) on Data Set 1

With the parameter set below, the calculation takes under a minute and uses about 1GB of memory.

```python
# Load data
data = np.loadtxt('../data/dataset1.csv', delimiter=',')

# Create Complex
max_edge_length = 1.5
max_dimension = 2
rips_complex = gd.RipsComplex(points=data, max_edge_length=max_edge_length)
simplex_tree = rips_complex.create_simplex_tree(max_dimension=max_dimension+1)

# Persistence
persistence = simplex_tree.persistence()
print(len(persistence))
```

```python
plot = gd.plot_persistence_barcode(persistence)
plot.show()
```

```python
plot = gd.plot_persistence_diagram(persistence)
plot.show()
```
